```python
jason = pd.read_csv("./data/jdf.csv")
new_df1 = pd.read_csv("./data/new_df1.csv")
new_df2 = pd.read_csv("./data/new_df2.csv")
```

```python
jason = jason.reset_index().drop(["index", "Unnamed: 0"], axis=1)
```

```python
# Remove beers that don't intersect for all 3 datasets
names = set(new_df1.beer_name) & set(new_df2.beer_name) & set(jason.name)
jason = jason[jason.name.isin(names)]
```

```python
# Merge it all together
new_df = new_df.join(new_df1[['beer_name','Brewery', 'ABV', 'Min IBU', 'Max IBU', 'Astringency',
       'Body', 'Alcohol', 'Bitter', 'Sweet', 'Sour', 'Salty', 'Fruits',
       'Hoppy', 'Spices', 'Malty']].set_index('beer_name'), on='name')
new_df = new_df.join(new_df2[['beer_name', 'beer_style']].set_index('beer_name'), on='name')
```
