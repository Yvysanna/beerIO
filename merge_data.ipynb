```python
import pandas as pd
import os
import json
```

```python
df1 = pd.read_csv('./data/beer_data_set.csv')
df2 = pd.read_csv('./data/beer_reviews.csv')
```

```python
df1.rename(columns={"Name":"beer_name"}, inplace=True)
names = set(df1.beer_name).intersection(set(df2.beer_name))
df_merged = pd.merge(df1, df2, on="beer_name", how="inner")
```

```python
new_df1 = df1[df1["beer_name"].isin(names)]
new_df2 = df2[df2["beer_name"].isin(names)]
new_df1.to_csv("/data/new_df1.csv")
new_df2.to_csv("/data/new_df2.csv")
```

```python
# Specify the folder containing the JSON files
folder_path = './data/beer-database'

# Create an empty list to store JSON data
merged_data = []

# Loop through each JSON file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            json_data = json.load(file)
            for i, data in enumerate(json_data["data"]):
                merged_data.append(data)

# Merge the data from all JSON files (example: merging into a list)
merged_result = merged_data

# Write the merged data to a new JSON file
output_file_path = "merged.json"
with open(output_file_path, 'w') as output_file:
    json.dump(merged_result, output_file, indent=4)
```

```python
jason = pd.read_json("./data/merged.json")
new_df1 = pd.read_csv("./data/new_df1.csv")
new_df2 = pd.read_csv("./data/new_df2.csv")
```

```python
names = set(new_df1.beer_name) & set(new_df2.beer_name) & set(jason.name)
```

```python
jdf = jason[jason["name"].isin(names)]
len(jdf)
```

```python
def extract_hex(json_str):
    try:
        return json_str['hex']
    except:
        return None
```

```python
def extract_glass(json_str):
    try:
        return json_str['name']
    except:
        return None
```

```python
# Add hex color
jdf.loc['hex_color'] = jdf['srm'].apply(extract_hex)

# Remove useless info about glass, just keep glass name
jdf.loc['glass'] = jdf.loc['glass'].apply(extract_glass)
```

```python
def extract_icon(json_str):
    try:
        data = json.loads(json_str)
        return data['icon']
    except:
        return None
```

```python

jdf.loc['icon'] = jdf['labels'].apply(extract_icon)
```

```python
jdf = jdf.reset_index().drop("index", axis=1)
```

```python
jdf.to_csv('./data/jdf.csv')
```
