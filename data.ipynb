```python
import pandas as pd
colors = dict()
df = pd.DataFrame({'color':[], 'freq':[], 'precision':[], 'recall':[], 'f1-score':[], 'support':[]})
with open('color.txt', 'r') as color:
    lines = color.readlines()
    for i, l in enumerate(lines[1:]):
        try:
            key, value = l.split('   ')
            df.loc[i] = [key] + [float(value.strip('\n'))] + [0.0] + [0.0] + [0.0] + [0]
        except ValueError:
            continue
```

```python
#df
with open('accuracy.txt', 'r') as acc:
    for l in acc.readlines():
        line = l.split('      ')
        if not any(c.isalnum() for c in line):
            continue
        if len(line) == 6:
            add_in = df.index[df['color'] == line[1]] 
            # print(line[2])
            df['precision'].iloc[add_in] += round(float(line[2].strip(' '))*100, 2)
            df['recall'].iloc[add_in] += round(float(line[3]), 2)
            df['f1-score'].iloc[add_in] += round(float(line[4]), 2)
            df['support'].iloc[add_in] += int(line[5].strip(' \n'))
#with open('results.txt', 'w') as infile:
```
